# Google Earth Engine Explorer

Google Earth Engine Explorer (EE Explorer) 是一个轻量级的地理空间图像数据查看器，可访问 Earth Engine 数据目录中的大量全球和区域数据集。它允许快速查看数据，能够在地球上任何地方缩放和平移，调整可视化设置，并分层数据以检查随时间变化。本教程将介绍 EE Explorer 应用程序的使用，包括：

- 如何在数据目录中查找数据
- 将数据添加到工作区
- 解释界面功能
- 如何定制数据可视化

本教程的目标是使您能够使用 EE Explorer，激发您发现和查看新数据，并提供一个起点，让您想象如何使用其他更强大的 Earth Engine 平台工具来回答有关地球当前状态和持续变化的问题。

## 内容

### Earth Engine Explorer

按照此链接 (https://explorer.earthengine.google.com) 进入 EE Explorer 应用程序。EE Explorer 由集成的数据目录和工作区组成。工作区是您查看数据的地方，数据目录是您发现并将数据导入工作区的地方。这两个组件都有一组按钮，用于在数据目录和工作区之间切换，并且有一个搜索栏，您可以通过关键字和位置名称查找数据集和位置。

按照上面的链接进入工作区，如下图所示。这是我们在本课程中花费大部分时间的地方，但在开始之前，我们需要一些数据。让我们看看数据目录。

### 数据目录

数据目录列出了可供在 Earth Engine 中查看和分析的数据集。EE Explorer 中可用于显示的数据是其中的一个子集。点击 EE Explorer 应用程序右上角的“数据目录”按钮。

在“数据目录”页面上，您将看到一列流行标签，链接到应用了这些标签的数据集。下面是各种数据类型和多天镶嵌图像的列表，包括对少数可用数据集的简要描述和直接链接。要访问其他数据集，请使用页面顶部的搜索栏。

点击几个流行标签以查看它们包含哪些类型的数据集。例如，`toa` 引出了一列描述“大气顶部反射率”的数据集。点击 `usgs` 引出了来自 USGS 的一系列数据集，包括 Landsat、MODIS 和从它们衍生出来的产品。

### 工作区

工作区是您可以添加、管理和查看图层以及调整可视化参数的地方。

#### 管理数据层

##### 添加数据层

要添加新图层，请单击左侧面板底部的“添加图层”按钮。这会打开一个对话框，在其中您可以搜索并选择要添加到工作区中的图层。

##### 图层可见性

每个图层旁边都有一个眼睛图标，单击该图标可以切换图层在工作区中是否可见。

##### 调整数据层日期

对于具有时间序列信息（例如 Landsat 影像）的多波段图像，可以使用时间滑块调整显示日期。

##### 添加多个图层

可以通过重复上述步骤来添加多个图层。

##### 重复数据集

如果您多次添加相同的图层，则会创建多个副本。

##### 重新排序图层

要更改图层的顺序，只需拖动它们即可。

##### 删除图层

要删除图层，只需单击图层右侧的垃圾桶图标即可。

#### 设置可视化参数

##### 数据带显示

###### 单波段灰度

要将单个波段显示为灰度图像，请在“波段”下拉菜单中选择一个波段。

###### 三波段真彩色

要将三个波段显示为真彩色图像，请在“波段”下拉菜单中选择三个波段。

###### 三波段假彩色

要将三个波段显示为假彩色图像，请在“波段”下拉菜单中选择三个波段。

##### 对比度、亮度和不透明度

###### 数据范围

可以使用“最小值”和“最大值”滑块调整数据范围。

###### Gamma

可以使用“Gamma”滑块调整 Gamma 值。

###### 不透明度

可以使用“不透明度”滑块调整图层的不透明度。

###### 调色板

对于分类数据，可以使用“调色板”选项卡为每个类别分配颜色。

#### 随时间变化的可视化

EE Explorer 允许您快速查看随时间变化的数据。只需添加多个时间序列数据集并使用时间滑块即可。

### 注意事项

- EE Explorer 是一个轻量级的查看器，旨在快速查看和检查数据。
- 它不提供分析功能，但可以用作探索数据并确定进一步分析所需数据的工具。
- EE Explorer 中可用的数据集是 Earth Engine 数据目录中可用数据集的子集。
- 有关更多信息，请参阅 [Google Earth Engine 文档](https://developers.google.com/earth-engine/)。
