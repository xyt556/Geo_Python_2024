```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
data = pd.read_csv('./data/Titanic.csv')

# Preview the dataset
print(data)
```

# **1. Countplot of passenger survival:**
   - This plot displays the count of passengers who survived and those who did        not survive the Titanic disaster.
   - It provides a visual representation of the survival distribution in the          dataset.


```python
# Exploratory Data Analysis

# Countplot of passenger survival
sns.countplot(x='Survived', data=data)
plt.title('Passenger Survival Count')
plt.show()
```

# **2. Countplot of survival by gender:**
   - This plot shows the count of passengers who survived and those who did not      survive, categorized by gender.
   - It helps to analyze the survival rate based on gender.

```python
# Countplot of survival by gender
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival Count by Gender')
plt.show()
```

# **3. Countplot of survival by passenger class:**
   - This plot illustrates the count of passengers who survived and those who did not survive, categorized by passenger class.
   - It allows for understanding the survival rate based on the passenger class.

```python
# Countplot of survival by passenger class
sns.countplot(x='Survived', hue='Pclass', data=data)
plt.title('Survival Count by Passenger Class')
plt.show()
```

# **4. Countplot of survival by number of siblings/spouses aboard:**
   - This plot presents the count of passengers who survived and those who did not survive, categorized by the number of siblings/spouses they had aboard.
   - It helps to explore the influence of having siblings/spouses on survival.

```python
# Countplot of survival by number of siblings/spouses aboard
sns.countplot(x='Survived', hue='SibSp', data=data)
plt.title('Survival Count by Siblings/Spouses Aboard')
plt.legend(title='SibSp')
plt.show()
```

# **5. Countplot of survival by number of parents/children aboard:**
   - This plot showcases the count of passengers who survived and those who did not survive, categorized by the number of parents/children they had aboard.
   - It provides insights into the impact of having parents/children on survival.


```python
# Countplot of survival by number of parents/children aboard
sns.countplot(x='Survived', hue='Parch', data=data)
plt.title('Survival Count by Parents/Children Aboard')
plt.legend(title='Parch')
plt.show()
```

# **6. Histogram of passenger age:**
   - This plot visualizes the distribution of passenger ages.
   - It allows for understanding the age composition of the passengers in the dataset.

```python
# Histogram of passenger age
sns.histplot(data['Age'].dropna(), kde=False)
plt.title('Passenger Age Distribution')
plt.show()
```

# **7. Boxplot of passenger fare by passenger class:**
   - This plot displays the distribution of passenger fares across different passenger classes using quartiles.
   - It provides insights into the fare differences among different passenger classes.

```python
# Boxplot of passenger fare by passenger class
sns.boxplot(x='Pclass', y='Fare', data=data)
plt.title('Passenger Fare by Passenger Class')
plt.show()
```

# **8. Heatmap of correlation between numeric features:**
   - This plot shows the correlation between numeric features using colors, where darker colors represent stronger correlations.
   - It helps to identify relationships and dependencies among numeric variables in the dataset.


```python
# Heatmap of correlation between numeric features
numeric_features = ['Age', 'Fare', 'SibSp', 'Parch']
correlation_matrix = data[numeric_features].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

# **9. Pairplot of numeric features:**
   - This plot creates scatterplots of all combinations of numeric features, showcasing the relationships between them.
   - It helps to visualize the pairwise interactions and identify potential correlations or patterns.


```python
# Pairplot of numeric features
sns.pairplot(data[numeric_features].dropna())
plt.title('Pairplot of Numeric Features')
plt.show()
```

# **10. Violinplot of passenger age by passenger class and survival:**
    - This plot depicts the distribution of passenger ages by passenger class and survival status using violin shapes.
    - It provides insights into the age distribution and its relationship with passenger class and survival.


```python
# Violinplot of passenger age by passenger class and survival
sns.violinplot(x='Pclass', y='Age', hue='Survived', data=data, split=True)
plt.title('Passenger Age by Passenger Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

# **11. Swarmplot of passenger fare by passenger class and survival:**
    - This plot displays individual data points of passenger fares, categorized by passenger class and survival status, along the categorical axis.
    - It helps to examine the fare distribution and its relationship with passenger class and survival.

```python
# Swarmplot of passenger fare by passenger class and survival
sns.swarmplot(x='Pclass', y='Fare', hue='Survived', data=data)
plt.title('Passenger Fare by Passenger Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

# **12. Pairwise correlation plot:**
    - This plot presents a heatmap of pairwise correlations between all numeric features, excluding the "PassengerId" column.
    - It provides a comprehensive view of the correlation matrix and identifies strong and weak correlations between variables.
