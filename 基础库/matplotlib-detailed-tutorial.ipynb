```python
plt.subplot(1,2,1)  # 这是同一行的两列图中的第一个图
plt.plot(x,y,"r--")  # 在第一个图中绘制x和y的关系，线型为红色虚线
plt.subplot(1,2,2)  # 这是同一行的两列图中的第二个图
plt.plot(y, x, 'g*-')  # 在第二个图中绘制y和x的关系，线型为绿色星号线
```

```python
# 指定行列
plt.subplot(2,2,1)  # 这是2行2列图中的第一个图
plt.plot(x,y,"r--")  # 在第一个图中绘制x和y的关系，线型为红色虚线
plt.subplot(2,2,2)  # 这是2行2列图中的第二个图
plt.plot(x,y,"g+")  # 在第二个图中绘制x和y的关系，线型为绿色加号
plt.subplot(2,2,3)  # 这是2行2列图中的第三个图
plt.plot(x,y,"r--")  # 在第三个图中绘制x和y的关系，线型为红色虚线
plt.subplot(2,2,4)  # 这是2行2列图中的第四个图
plt.plot(y,x,"y+")  # 在第四个图中绘制y和x的关系，线型为黄色加号
```

## 3. Setting Colors, Linewidths, Linetypes

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where 'b' means blue, 'g' means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
fig, axes = plt.subplots() # If we do not specify number of rows and columns, it will create a single plot figure
axes.plot(x,x**2, "b.-")
axes.plot(x,x**3, "g--")
axes.plot(x,x,"r+") # we can add any number of lines inside a single plot
```

在matplotlib中，`marker`参数用于设置数据点的形状。在你的代码`marker="<"`中，`"<"`表示数据点的形状是一个向左的三角形¹。

以下是一些可用的预定义标记¹：
- `"."`：点
- `","`：像素
- `"o"`：圆圈
- `"v"`：向下的三角形
- `"^"`：向上的三角形
- `"<"`：向左的三角形
- `">"`：向右的三角形
- `"1"`：向下的三叉戟
- `"2"`：向上的三叉戟
- `"3"`：向左的三叉戟
- `"4"`：向右的三叉戟
- `"8"`：八边形
- `"s"`：正方形
- `"p"`：五边形
- `"P"`：加号（实心）
- `"*"`：星形
- `"h"`：六边形1
- `"H"`：六边形2
- `"+"`：加号
- `"x"`：x
- `"X"`：x（实心）
- `"D"`：菱形
- `"d"`：窄菱形
- `"|"`：垂直线
- `"_"`：水平线

你可以通过`plt.plot([1, 2, 3], marker='<')`来查看`"<"`标记的样子。

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x=x, y=y,marker="d",s=100,cmap="autumn",color = '#ff0000')#已经明确设置了点的颜色（通过color参数），cmap参数就会被忽略
```

```python
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)
```
