<h2 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">üõ†Importing Necessary Libraries‚ú®</h2>


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

<h2 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">üìÅ Sample Datasets üìÅ </h2>


```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')  
exercise = sns.load_dataset('exercise')  
titanic = sns.load_dataset('titanic')  
fmri = sns.load_dataset('fmri')
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Scatter Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A scatter plot showing the relationship between sepal length and sepal width for different iris flower species. The plot can help identify potential patterns or clusters among iris species based on these two attributes.</p>

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Line Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A line plot depicting the fMRI signal response over time for a specific subject. This plot helps visualize the time-dependent changes in the subject's brain activity captured through fMRI scans.</p>

```python
sns.lineplot(x='timepoint', y='signal', data=fmri)
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Bar Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information :  A bar plot illustrating the total bill amount for each day of the week based on customer tips. The plot allows us to compare the total bill amounts on different days and identify any significant </p>

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Violin Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A violin plot visualizing the distribution of data, particularly showcasing the kernel density estimation along with the box plot. It provides insights into the data's density, skewness, and presence of multiple modes.</p>

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Histogram Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A histogram displaying the distribution of total bill amounts in the sample dataset. This plot gives an overview of how the total bill amounts are distributed and helps identify common ranges.</p>

```python
sns.histplot(x='total_bill', data=tips)
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Pair Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A pair plot showing the relationship between various attributes of iris flowers, colored by species. This plot enables us to visualize the correlations between different attributes and how they vary across different iris species.</p>

```python
sns.pairplot(iris, hue='species')
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Heat Map üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A heatmap visualizing the correlation between different attributes of iris flowers. This heatmap provides a color-coded representation of attribute correlations, making it easier to identify strong or weak relationships.</p>

```python
sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Joint Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A joint plot highlighting the relationship between total bill and tip amount in the sample dataset. This plot combines a scatter plot and histograms to show the distribution of both variables and their correlation.</p>

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Regression/Reg Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information :  A regression plot displaying the linear relationship between total bill and tip in the dataset. This plot helps visualize the trend and strength of the linear relationship between the two variables.</p>

```python
sns.regplot(x='total_bill', y='tip', data=tips)
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Categorical Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A count plot representing the frequency of categories in the sample dataset. This plot displays the occurrence of each category, helping identify the most and least common categories.</p>

```python
sns.catplot(x='sex', y='total_bill', hue='time', data=tips, kind='bar')
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Distribution Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A distribution plot illustrating the spread of data across a single variable. It provides valuable insights into the data's central tendency, variability, and skewness.</p>

```python
sns.displot(tips['total_bill'], kde=True)
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Swarm Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information :A swarm plot visualizing the distribution of scores for different students. This plot shows individual data points and their distribution, making it useful for comparing scores across students.</p>

```python
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
scores = np.random.randint(50, 100, size=len(students))
sns.swarmplot(x=students, y=scores)
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä FaceGrid Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A facet grid with scatter plots of total bill and tip amount, grouped by gender and meal time. This grid allows us to examine how total bill and tip amount vary based on both gender and meal time.</p>

```python
g = sns.FacetGrid(tips, col='sex', row='time', margin_titles=True)
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip')
g.set_axis_labels('Total Bill ($)', 'Tip ($)')
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Kde Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A KDE plot displaying the estimated probability density function of a random variable. This plot provides a smooth representation of the underlying probability distribution.</p>

```python
data = np.random.normal(loc=0, scale=1, size=1000)

sns.kdeplot(data)
plt.show()
```

<h4 style="text-align: center; font-size: 32px; background-color: #ffffcc; padding: 10px; border-radius: 10px;"> üìàüìä Point Plot üìàüìä </h4>

<p style="text-align: left; font-size: 20px; background-color: #ffffcc; padding: 10px; border-radius: 10px;">Information : A point plot illustrating data points on a numeric scale. This plot is useful for visualizing individual data points and their positions on a numeric axis.</p>

```python
x_vals = [1, 2, 3, 4, 5]
y_vals = [2, 4, 1, 3, 5]

sns.pointplot(x=x_vals, y=y_vals)
plt.show()
```
