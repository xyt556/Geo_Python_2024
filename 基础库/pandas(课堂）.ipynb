Pandas 可以读取多种文件格式，包括但不限于以下几种：

    CSV（逗号分隔值）文件：Pandas 提供了 read_csv() 函数用于读取 CSV 文件，并将其转换为 DataFrame 对象。

    TXT（文本）文件：可以使用 read_csv() 函数的 sep 参数指定分隔符为一个制表符或空格，将 TXT 文件读取为 DataFrame 对象。

    Excel 文件：Pandas 提供了 read_excel() 函数用于读取 Excel 文件，并将其转换为 DataFrame 对象。

    SQL 数据库：Pandas 提供了 read_sql() 函数用于直接从 SQL 数据库中读取数据，并将其转换为 DataFrame 对象。

    JSON 文件：可以使用 read_json() 函数将 JSON 文件读取为 Pandas 的 DataFrame 对象。

    HTML 表格：可以使用 read_html() 函数从 HTML 文件中提取表格数据，并将其转换为 DataFrame 对象。

`pd.set_option()`是Pandas库中的一个函数，用于设置指定选项的值¹²。这个函数可以让你自定义Pandas的一些行为，例如更改显示的最大行数或列数²。

函数的语法如下：
```python
pandas.set_option(pat, value)
```
参数的含义如下：

`df.append()`是Pandas库中的一个函数，用于将其他DataFrame的行附加到给定DataFrame的末尾，返回一个新的对象。在其他DataFrame中不在调用者中的列将作为新列添加¹。

函数的语法如下：
```python
DataFrame.append(other, ignore_index=False, verify_integrity=False, sort=False)
```
参数的含义如下：
- `other`：DataFrame或Series/dict-like对象，或这些对象的列表。要附加的数据¹。
- `ignore_index`：布尔值，默认为False。如果为True，结果轴将被标记为0，1，…，n - 1¹。

- `sort`：布尔值，默认为False。如果self和other的列不对齐，则排序列¹。

这个函数返回一个DataFrame，包括调用者的行和其他的行¹。

## Pandas 绘图

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
air_quality = pd.read_csv("./data/air_quality_no2.csv", index_col=0, parse_dates=True)

air_quality.head()
````

````python
air_quality.plot()
plt.show()
````

````python
air_quality["station_paris"].plot()
plt.show()
````

````python
air_quality.plot.scatter(x="station_london", y="station_paris", alpha=0.5)
````

````python
air_quality.plot.box()
plt.show()
````

````python
# 每一列都在一个单独的子图中。
axs = air_quality.plot.area(figsize=(12, 4), subplots=True)
````

````python
# 创建一个新的图表对象fig和一个子图对象axs，设置图表的大小为12x4
fig, axs = plt.subplots(figsize=(12, 4))

# 使用空气质量数据创建一个面积图，ax=axs表示在刚刚创建的子图上绘制
air_quality.plot.area(ax=axs)

# 设置y轴的标签为"NO$_2$ concentration"
axs.set_ylabel("NO$_2$ concentration")

# 将图表保存为一个名为"no2_concentrations.png"的图片文件，保存在"./tmp/"目录下
fig.savefig("./tmp/no2_concentrations.png")

# 显示图表
plt.show()
````
