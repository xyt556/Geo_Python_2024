## 1. Figure and Subplots

#### Plots in matplotlib reside in a figure object

```python
fig = plt.figure()
plt.show()
```

```python
# Creating four subplots in the fig 
p1 = fig.add_subplot(2,2,1)
p2 = fig.add_subplot(2,2,2)
p3 = fig.add_subplot(2,2,3)
p4 = fig.add_subplot(2,2,4)
```

```python
# Plotiing inside one of the subplots
plt.plot(np.arange(10),'k--')
```

```python
# Plotting inside a specific subplot
p1.scatter(np.arange(30),np.arange(30) + np.random.randn(30))
```

```python
# Creating multiple subplots at once
fig,axes = plt.subplots(2,1)
axes
```

```python
# Accessing the subplots and plotting inside them
axes[0].plot(np.arange(10))
axes[1].plot(np.arange(10))
```

## 2. Simple Plot

```python
# A simple plot in matplotlib
fig,p = plt.subplots(1,1)
x = np.arange(10)
y = np.arange(10)*2
plt.plot(x,y,'g*')
```

```python
fig,p = plt.subplots(1,1)
plt.plot(np.random.randn(30),'ko--',label = 'One')
```

```python
# Setting labels and legends
p.set_xlabel('X-axis')
p.set_ylabel('Y-axis')
p.set_title('MyPlot')
p.legend(loc='best')
```

```python
# Two plots inside a figure
plt.figure()
x = np.arange(10)
y = x ** 2
z = x ** 3
plt.plot(y,'o-',z,'-o')
plt.xlabel('Numbers')
plt.ylabel('Squares and Cubes')
plt.legend(['Squares','Cubes'])
```

## 3.Scatter Plot

```python
plt.figure()
x = np.arange(10)
y = np.arange(10) * 3
plt.scatter(x,y,s = 100, c = 'green')
# Setting x - axis values
plt.xticks(np.arange(0,10,1))
# Setting y-axis values
plt.yticks(np.arange(0,30,3))
```

# 4.Bar Plot

```python
plt.figure()
x = np.arange(10)
y = x * 2
plt.bar(x,y,width = 0.8)
plt.xticks(x)
plt.yticks(y)
```

```python
plt.figure()
z = x * 4
plt.bar(x,y,width = 0.4,color = 'green')
plt.bar(x,z,width = 0.4,bottom = x,color = 'orange')
plt.xticks(x)
```

```python
plt.figure()
x = np.arange(10)
y = x * 2
plt.bar(x,y,width = 0.4,color = 'yellow')
plt.bar(x+0.5,y*2,width = 0.4,color = 'green')
```

## 5.Plotting a Histogram

```python
plt.figure()
plt.hist([1,2,3,4,5,2,4,7,8,4,5,2,1,8,8,0,9,0,9,2,5,2,1,4])
plt.xticks(range(10))
```

```python
plt.figure()
x = np.random.normal(100, 10, 300)
plt.hist(x)
```

## 6. PieChart

```python
plt.figure()
x = np.array([10,30,20,15,25])
plt.pie(x)
```

```python
plt.figure()
y = np.array([30,20,40,30,10])
food = ["Bananas","Small Bananas","Big Bananas","Tiny Bananas","Extra Large Bananas"]
plt.pie(y, labels=food)
```
