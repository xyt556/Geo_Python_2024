RandomForest算法的简要解释源自其名称。与仅利用单个决策树的预测不同，该算法将大量决策树的集成结果（即一片森林）作为输出。名称中的“随机”部分来自术语“bootstrap aggregating”或“bagging”。这意味着森林中的每棵树只能在完整训练数据集的某个子集上进行训练（子集由有放回抽样确定）。每棵树中未见的训练数据元素被保留为“out-of-bag”，用于估计准确性。随机性还有助于决定每棵决策树中每个节点处看到的特征输入变量。一旦所有单独的树都适应于训练数据的随机子集，并在每个节点处使用一组随机的特征变量，就会使用它们的集合来给出最终的预测结果。

在分类模式下，这意味着如果您使用500棵树预测5个类别，输出预测将是500棵树中预测它最多的类别。投票赞成获胜类别的树的数量比例可以用来诊断您的训练数据相对于图像其余部分的代表性。以500棵树为例，如果您的像素被475棵树中的475棵树投票为“森林”地覆盖类别，您可以说这是一个相对确定的预测。另一方面，如果您的像素获得250票“森林”和225票“灌木”，您可以将其解释为一个天生令人困惑的像素（也许它是一个混合像素，或者它是一个矮小的森林），或者作为您需要在这些类型的像素中获取更多的训练数据样本的指标。

最后，随机森林还具有其他一些优点：
+ 它为您提供了“变量重要性”的度量，该度量与您的输入特征（例如光谱波段）在分类中的有用程度有关
+ 每棵树中的“out-of-bag”样本可用于验证每棵树。将所有树的这些预测准确性进行分组，有时可以给出类似于交叉验证的无偏误差率的估计值。
+ 可用于回归、无监督聚类或有监督分类
+ 支持许多流行的编程语言，包括Python、R和MATLAB
+ 免费且开源，速度快

## 准备数据集
#### 打开图像
我们的第一步是回顾我们之前章节的内容，读取在第4章中创建的示例图像和ROI图像（链接到[网站](chapter_4_vector.html)或[Notebook](chapter_4_vector.ipynb)）：

````python
# 导入 Python 3 的 print 函数和除法
from __future__ import print_function, division

# 导入 GDAL、NumPy 和 matplotlib
from osgeo import gdal, gdal_array
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 告诉 GDAL 抛出 Python 异常，并注册所有驱动程序
gdal.UseExceptions()
gdal.AllRegister()

# 读取我们的图像和 ROI 图像
img_ds = gdal.Open('../../example/LE70220491999322EDC01_stack.gtif', gdal.GA_ReadOnly)
roi_ds = gdal.Open('../../example/training_data.gtif', gdal.GA_ReadOnly)

# 创建用于存储图像数据的数组
img = np.zeros((img_ds.RasterYSize, img_ds.RasterXSize, img_ds.RasterCount),
               gdal_array.GDALTypeCodeToNumericTypeCode(img_ds.GetRasterBand(1).DataType))
# 读取图像的每个波段数据到数组中
for b in range(img.shape[2]):
    img[:, :, b] = img_ds.GetRasterBand(b + 1).ReadAsArray()
    
# 读取 ROI 数据
roi = roi_ds.GetRasterBand(1).ReadAsArray().astype(np.uint8)

# 显示图像和 ROI 数据
plt.subplot(121)
plt.imshow(img[:, :, 4], cmap=plt.cm.Greys_r)
plt.title('SWIR1')

plt.subplot(122)
plt.imshow(roi, cmap=plt.cm.Spectral)
plt.title('ROI Training Data')

plt.show()
````

#### 将Y与X配对
现在我们已经有了要分类的图像（我们的X特征输入）和带有土地覆盖标签的ROI（我们的Y标记数据），我们需要将它们配对到NumPy数组中，以便将它们提供给随机森林：

````python
# 找出有多少个非零条目 -- 即有多少个训练数据样本？
n_samples = (roi > 0).sum()
print('我们有 {n} 个样本'.format(n=n_samples))

# 我们的分类标签是什么？
labels = np.unique(roi[roi > 0])
print('训练数据包括 {n} 个类别: {classes}'.format(n=labels.size, 
                                              classes=labels))
# 我们需要一个包含我们特征的 "X" 矩阵，和一个包含我们标签的 "y" 数组
#     这些将有 n_samples 行
#     在其他语言中，我们需要分配这些并循环填充它们，但是 NumPy 可以更快地完成

X = img[roi > 0, :]  # 包括第8波段，即Fmask
y = roi[roi > 0]

print('我们的 X 矩阵的大小为: {sz}'.format(sz=X.shape))
print('我们的 y 数组的大小为: {sz}'.format(sz=y.shape))

# 使用 Fmask 掩码掉云、云影和雪
clear = X[:, 7] <= 1

X = X[clear, :7]  # 我们现在可以放弃 Fmask 波段了
y = y[clear]

print('掩码后，我们的 X 矩阵的大小为: {sz}'.format(sz=X.shape))
print('掩码后，我们的 y 数组的大小为: {sz}'.format(sz=y.shape))
````

## 训练随机森林
现在我们已经有了要分类的图像（我们的X特征输入）和带有土地覆盖标签的ROI（我们的Y标记数据），我们需要将它们配对到NumPy数组中，以便将它们提供给随机森林：

````python
from sklearn.ensemble import RandomForestClassifier

# 使用 500 棵树初始化我们的模型
rf = RandomForestClassifier(n_estimators=500, oob_score=True)

# 将我们的模型拟合到训练数据上
rf = rf.fit(X, y)
````

#### 随机森林诊断

随着我们拟合的随机森林模型，我们可以查看“Out-of-Bag”（OOB）预测得分：

````python
print('Our OOB prediction of accuracy is: {oob}%'.format(oob=rf.oob_score_ * 100))
````

为了帮助我们了解哪些光谱波段很重要，我们可以查看特征重要性分数：

```python
# Get feature importances from the trained RandomForest model
feature_importances = forest.feature_importances_

# Assuming you have a list of feature names (e.g., band names)
feature_names = ["Band 1", "Band 2", "Band 3", ...]

# Create a dictionary to associate feature names with their importances
importance_dict = dict(zip(feature_names, feature_importances))

# Sort the dictionary by importance (descending order)
sorted_importance_dict = {k: v for k, v in sorted(importance_dict.items(), key=lambda item: item[1], reverse=True)}

# Print the sorted feature importances
for band, importance in sorted_importance_dict.items():
    print(f"{band}: {importance:.4f}")
```

这将使您了解哪些光谱波段对分类贡献最大。记住，更高的重要值表示对分类决策的影响更大。

````python
bands = [1, 2, 3, 4, 5, 7, 6]

for b, imp in zip(bands, rf.feature_importances_):
    print('Band {b} importance: {imp}'.format(b=b, imp=imp))
````

根据最大权重，看起来SWIR1和绿色波段对我们最有用。并不奇怪，热红外波段并不是很有用，因为如果您只在不受城市热岛影响的非常小的区域内查看，温度对土地覆盖类型的划分不是很明显。

让我们查看交叉表以查看类别混淆。为此，我们将导入[Pandas](http://pandas.pydata.org/)库以获取一些帮助：

````python
import pandas as pd

# 设置一个数据框架 -- 就像 R 一样
df = pd.DataFrame()
df['truth'] = y
df['predict'] = rf.predict(X)

# 交叉制表预测结果
print(pd.crosstab(df['truth'], df['predict'], margins=True))
````

难以置信，对吧？我非常怀疑真正的混淆矩阵会达到100%的准确率。实际情况可能是，我们在机器学习算法中使用了大量的树，以最好地找出训练数据中的模式。在有足够的信息和努力的情况下，该算法准确地学到了我们提供的内容。要求在训练数据上验证机器学习算法是一种无用的练习，会使准确率过高。

相反，我们可以采用交叉验证方法，在数据集的子集上进行训练，然后使用我们没有训练的部分进行预测和准确性评估。为了获得性能的稳健度量，我们可以多次使用随机子集运行这个过程。

## 预测整个图像

现在我们已经拟合了随机森林分类器，我们可以尝试对整个图像进行分类：

````python
# 使用完整图像，忽略 Fmask 波段，并将其重塑为长形式的二维数组 (nrow * ncol, nband) 用于分类
new_shape = (img.shape[0] * img.shape[1], img.shape[2] - 1)

img_as_array = img[:, :, :7].reshape(new_shape)
print('从 {o} 重塑为 {n}'.format(o=img.shape,
                              n=img_as_array.shape))

# 现在为每个像素预测
class_prediction = rf.predict(img_as_array)

# 重塑我们的分类地图
class_prediction = class_prediction.reshape(img[:, :, 0].shape)
````

````python
# 首先设置一个 5-4-3 复合图像
def color_stretch(image, index, minmax=(0, 10000)):
    colors = image[:, :, index].astype(np.float64)

    max_val = minmax[1]
    min_val = minmax[0]

    # 强制最大和最小值
    colors[colors[:, :, :] > max_val] = max_val
    colors[colors[:, :, :] < min_val] = min_val

    for b in range(colors.shape[2]):
        colors[:, :, b] = colors[:, :, b] * 1 / (max_val - min_val)
        
    return colors
    
img543 = color_stretch(img, [4, 3, 2], (0, 8000))

# 参见 https://github.com/matplotlib/matplotlib/issues/844/
n = class_prediction.max()
# 接下来为我们的分类地图设置一个颜色映射
colors = dict((
    (0, (0, 0, 0, 255)),  # 无数据
    (1, (0, 150, 0, 255)),  # 森林
    (2, (0, 0, 255, 255)),  # 水域
    (3, (0, 255, 0, 255)),  # 草地
    (4, (160, 82, 45, 255)),  # 裸土
    (5, (255, 0, 0, 255))  # 城市
))
# 将 0 - 255 转化为浮点数 0 - 1
for k in colors:
    v = colors[k]
    _v = [_v / 255.0 for _v in v]
    colors[k] = _v
    
index_colors = [colors[key] if key in colors else 
                (255, 255, 255, 0) for key in range(1, n + 1)]
cmap = plt.matplotlib.colors.ListedColormap(index_colors, 'Classification', n)

# 现在在图像旁边显示分类地图
plt.subplot(121)
plt.imshow(img543)

plt.subplot(122)
plt.imshow(class_prediction, cmap=cmap, interpolation='none')

plt.show()
````
